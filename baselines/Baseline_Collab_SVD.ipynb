```python
import pandas as pd
from surprise import Dataset, Reader

# Load datasets
df_train = pd.read_json('/content/drive/path/beeradvocate_train.json', lines=True)
df_val = pd.read_json('/content/drive/path/beeradvocate_val.json', lines=True)
df_test = pd.read_json('/content/drive/path/beeradvocate_test.json', lines=True)

# Check for NaNs and drop them
df_train_clean = df_train[['review/profileName', 'beer/beerId', 'review/overall']].dropna()
df_val_clean = df_val[['review/profileName', 'beer/beerId', 'review/overall']].dropna()
df_test_clean = df_test[['review/profileName', 'beer/beerId', 'review/overall']].dropna()

# Prepare data for Surprise
r = Reader(rating_scale=(1, 5))
train_data = Dataset.load_from_df(df_train_clean, r)
val_data = Dataset.load_from_df(df_val_clean, r)
test_data = Dataset.load_from_df(df_test_clean, r)

trainset = train_data.build_full_trainset()
valset = list(val_data.df.itertuples(index=False, name=None))
testset = list(test_data.df.itertuples(index=False, name=None))

print(f"Data cleaned and loaded.")
print(f"Trainset: {trainset.n_users} users, {trainset.n_items} items.")
print(f"Testset size: {len(testset)}")
```

```python
from surprise import SVD, accuracy

# Initialize and train SVD
algo = SVD()
algo.fit(trainset)

# Predict and evaluate on validation set
val_predictions = algo.test(valset)
print("Validation Set Performance:")
val_rmse = accuracy.rmse(val_predictions)
val_mae = accuracy.mae(val_predictions)

# Predict and evaluate on test set
test_predictions = algo.test(testset)
print("\nTest Set Performance:")
test_rmse = accuracy.rmse(test_predictions)
test_mae = accuracy.mae(test_predictions)
```
